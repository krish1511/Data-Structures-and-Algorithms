# Bubble Sort

## Bubble sort using for loop 

```python
def bubbleSort(nums):
    for i in range(len(nums)-1 , 0 , -1):
        for j in range(i):
            if nums[j] > nums[j+1]:
                nums[j] , nums[j+1] = nums[j+1] , nums[j]
    print(nums)
                
nums=[6,4,2,7,1,3]
bubbleSort(nums)
```

## bubble sort using while loop

```python
def bubbleSort(nums):
    i = 0
    while i < len(nums)-1:
        j = 0
        while j < len(nums)-1:
            if nums[j] > nums[j+1]:
                nums[j] , nums[j+1] = nums[j+1] , nums[j]
            j+=1
        i+=1
        
    print(nums)
        
        
nums=[3,9,4,2,1,5]
bubbleSort(nums)
```

# Selection Sort 

## Selection Sort using for loop

```python
def selectionSort(nums):
    for i in range(len(nums)):
        minPos = i
        for j in range(i,len(nums)):
            if nums[j] < nums[minPos]:
                minPos = j
        nums[i] , nums[minPos] = nums[minPos] , nums[i]
    print(nums)
nums = [9,3,1,2,7,4]
selectionSort(nums)
```

## Selection Sort using while loop 

```python
def selectionSort(nums):
    i = 0
    while i < len(nums):
        minPos = i
        for j in range(i , len(nums)):
            if nums[j] < nums[minPos]:
                minPos = j
        nums[i] , nums[minPos] = nums[minPos] , nums[i]
            
        i+=1
    print(nums)
nums = [9,6,4,7,2,1]
selectionSort(nums)
```

# InsertionSort

```python
def insertSort(nums):
    for i in range(1, len(nums)):
        cur_ele = nums[i]
        pos = i
        while cur_ele < nums[pos - 1] and pos > 0:
            nums[pos] = nums[pos - 1]
            pos = pos - 1
        nums[pos] = cur_ele
    print(nums)
nums = [6,8,1,2,3,7]
insertSort(nums)
```

# MergingSort

```python
def mergeSort(lst):
    # Dividing the list
    if len(lst) > 1:
        mid = len(lst) // 2
        left = lst[:mid]
        right = lst[mid:]
        # dividing the list into two[left and right]
        mergeSort(left)
        mergeSort(right)
        
        # merging the compared values
        i=0
        j=0
        k=0
        while i < len(left) and j < len(right):
            if left[i] < right[j]:
                lst[k] = left[i]
                i+=1
                k+=1
            else:
                lst[k] = right[j]
                j+=1
                k+=1
        # whether any values are left out
        while i < len(left):
            lst[k] = left[i]
            i+=1
            k+=1
        while j < len(right):
            lst[k] = right[j]
            j+=1
            k+=1
     
    
nums = int(input('enter the elments you want in list'))
lst = [int(input()) for i in range(nums)]
mergeSort(lst)
print(lst)
```

# CountingSort

```python
def countingsort(lst, max):
    count = [0] * (max + 1)       #7    len(lst)+1 zeros
    
    for i in lst:
        count[i] +=1       # it count the repeated(incremented) values
        
    start = 0                 # list starting from 0
    for i in range(len(count)):       # empty list has to be given of max+1 times
        for j in range(count[i]):    # count[i] = according the element count, it increases the count of repeated values
            lst[start] = i           #lst[2] = 2 times
            start += 1
    return lst
lst = [2,9,6,5,2,3]
countingsort(lst, max(lst))
            
    
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None
        self.height = 1
class AVL:
    def getHeight(self, root):
        if not root:
            return 0
        return root.height
    
    def getBalance(self, root):
        if not root:
            return 0
        return self.getHeight(root.left) - self.getHeight(root.right)
    def rightRotate(self, z):
        y = z.left
        T3 = y.right
        # right rotate
        y.right = z
        z.left = T3
        
        z.height = 1 + max(self.getHeight(z.left) , self.getHeight(z.right))
        y.height = 1 + max(self.getHeight(y.left) , self.getHeight(y.right))
        
        return y
    def leftRotate(self, z):
        y = z.right
        T2 = y.left
        # left rotate
        y.left = z
        z.right = T2
        
        z.height = 1 + max(self.getHeight(z.left) , self.getHeight(z.right))
        y.height = 1 + max(self.getHeight(y.left) , self.getHeight(y.right))
        
        return y
    def insert(self, root, data):
        if not root:
            return Node(data)
        elif data < root.data:
            if root.left is None:
                root.left = Node(data)
            else:
                root.left =  self.insert(root.left,data)
        else:
            if root.right is None:
                root.right = Node(data)
            else:
                root.right = self.insert(root.right,data)
        # updating heights
        root.height = 1 + max(self.getHeight(root.left) , self.getHeight(root.right))
        # balancing Nodes
        balance = self.getBalance(root)
        # unbalancing nodes
        if balance > 1 and data < root.left.data:
            return self.rightRotate(root)
        if balance < -1 and data > root.right.data:
            return self.leftRotate(root)
        if balance > 1 and data > root.left.data:
            root.left = self.leftRotate(root.left)
            return self.rightRotate(root)
        if balance < -1 and data < root.right.data:
            root.right = self.rightRotate(root.right)
            return self.leftRotate(root)
        return root
    
    def delete(self, root,data):
        if not root:
            return root
        elif data < root.data:
            root.left = self.delete(root.left, data)
        elif data > root.data:
            root.right = self.delete(root.right, data)
        else:
            if root.left is None:
                temp = root.right
                root = None
                return temp
            elif root.right is None:
                temp = root.left
                root = None
                return temp
            
            temp = self.inOrderSuccessor(root.right)
            root.data = temp.data
            root.right = self.delete(root.right, temp.data)
        if root is None:
            return root
        
        root.height =  1 + max(self.getHeight(root.left) , self.getHeight(root.right))
        
        balance = self.getBalance(root)
        
        if balance > 1 and self.getBalance(root.left) >= 0:
            return self.rightRotate(root)
        if balance < -1 and self.getBalance(root.right) <= 0:
            return self.leftRotate(root)
        if balance > 1 and self.getBalance(root.left) < 0:
            root.left = self.leftRotate(root.left)
            return self.rightRotate(root)
        if balance < -1 and self.getBalance(root.right) > 0:
            root.right = self.rightRotate(root.right)
            return self.leftRotate(root)
        return root
    
    def inOrderSuccessor(self, root):
        myVal = root
        while myVal.left is not None:
            myVal = myVal.left
        return myVal
            
    def preOrder(self, root):
        if not root:
            return 0
        print('{0}'.format(root.data),end=' ')
        self.preOrder(root.left)
        self.preOrder(root.right)
avt = AVL()
root = None
nums = [9, 5, 10, 0, 6, 11, -1, 1, 2] 

for num in nums:
    root = avt.insert(root,num)
    
avt.preOrder(root)
print()
data = 10
root = avt.delete(root, data)

avt.preOrder(root)
            
```

## CircularQueue

```python
class CircularQueue:
    def __init__(self, size):
        self.size = size
        self.queue = [None for i in range(size)]
        self.rear = self.front = -1
    def enqueue(self, data):
        if (self.rear + 1) % self.size == self.front:
            print('queue is full')
            return
        elif self.rear == -1 and self.front == -1:
            self.rear = 0
            self.front = 0
            self.queue[self.rear] = data
        else:
            self.rear = (self.rear + 1) % self.size
            self.queue[self.rear] = data
    def dequeue(self):
        if self.rear == -1 and self.front == -1:
            print('queue is empty')
            return
        elif self.rear == self.front:
            temp = self.queue[self.front]
            self.rear = -1
            self.front = -1
            return temp
        else:
            temp = self.queue[self.front]
            self.front = (self.front + 1) % self.size
            return temp
    def display(self):
        if (self.rear + 1) % self.size == self.front:
            print('queue is full')
            return
        if self.rear == -1 and self.front == -1:
            print('queue is empty')
            return
        elif self.rear >= self.front:
            print('elements in the circularqueue are :',end=' ')
            for i in range(self.front , self.rear + 1):
                print(self.queue[i],end=' ')
            print()
        else:
            print('elements in the circularqueue are :',end=' ')
            for i in range(self.front , self.size):
                print(self.queue[i],end=' ')
            for i in range(0, self.rear + 1):
                print(self.queue[i],end=' ')
            print()
            
cq = CircularQueue(6)
cq.enqueue(10)
cq.enqueue(20)
cq.enqueue(30)
cq.enqueue(40)
cq.display()
print('delete elements:',cq.dequeue())
print('delete elements:',cq.dequeue())
cq.display()
cq.enqueue(50)
cq.enqueue(60)
cq.display()
            
```

## BinarySearchTree [BST]

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None
class BST:
    def __init__(self, root = None):
        self.root = root
    def insert(self, data):
        if self.root is None:
            self.root = Node(data)
        else:
            self._insert(self.root, data)
    def _insert(self, cur_node, data):
        if data < cur_node.data:
            if cur_node.left is None:
                cur_node.left = Node(data)
            else:
                self._insert(cur_node.left, data)
        else:
            if cur_node.right is None:
                cur_node.right = Node(data)
            else:
                self._insert(cur_node.right, data)
                
    def delete(self, cur_node, data):
        if cur_node is None:
            return cur_node
        elif data < cur_node.data :
            cur_node.left = self.delete(cur_node.left, data)
        elif data > cur_node.data:
            cur_node.right = self.delete(cur_node.right, data)
        else:
            if cur_node.left is None:
                temp = cur_node.right
                cur_node = None
                return temp
            elif cur_node.right is None:
                temp = cur_node.left 
                cur_node = None
                return temp
            else:
                temp = self.inOrderSuccessor(cur_node.right)
                cur_node.data = temp.data
                cur_node.right = self.delete(cur_node.right, temp.data)
            return cur_node
        
    def inOrderSuccessor(self, cur_node):
        myVal = cur_node
        while myVal.left is not None:
            myVal = myVal.left
        return myVal
    def search(self, cur_node, data):
        if cur_node is None:
            print('data not found')
            return False
        
        elif cur_node.data == data :
            print(data found)
            return True
        
        elif data < cur_node.data :
            cur_node.left = self.search(cur_node.left, data)
        else:
            cur_node.right = self.search(cur_node.right,data)
            
    def inOrder(self, cur_node):
        if cur_node:
            self.inOrder(cur_node.left)
            print('{0}'.format(cur_node.data),end= ' ')
            self.inOrder(cur_node.right)
    def preOrder(self, cur_node):
        if cur_node:
            print('{0}'.format(cur_node.data),end= ' ')
            self.preOrder(cur_node.left)
            self.preOrder(cur_node.right)
    def postOrder(self, cur_node):
        if cur_node:
            self.postOrder(cur_node.left)
            self.postOrder(cur_node.right)
            print('{0}'.format(cur_node.data),end= ' ')
    def print(self):
        if self.root is not None:
            self._print(self, cur_node)
    def _print(self, cur_node):
        if cur_node:
            self._print(self, cur_node.left)
            print(cur_node.data)
            self._print(self, cur_node.right)
    
```

## CircularSinglyLinkedList 

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
class CircularSinglyLinkedList:
    def __init__(self):
        self.head = None
    def insertHead(self, newNode):
        lastNode = self.head
        while lastNode.next is not self.head:
            lastNode = lastNode.next
        temporaryNode = self.head
        self.head = newNode
        newNode.next = temporaryNode
        lastNode.next = self.head
        del temporaryNode
    def insertAt(self, newNode, position):
        currentNode = self.head
        currentPos = 0
        while True:
            if currentPos == position:
                previousNode.next = newNode
                newNode.next = currentNode
                currentNode.next = self.head
                break
            previousNode = currentNode
            currentNode = currentNode.next
            currentPos += 1
        
    def insertEnd(self, newNode):
        if self.head is None:
            self.head = newNode
            newNode.next = self.head
            return
        lastNode = self.head
        while lastNode.next is not self.head:
            lastNode = lastNode.next
        lastNode.next = newNode
        newNode.next = self.head
    def deleteHead(self):
        lastNode = self.head
        while lastNode.next is not self.head:
            lastNode = lastNode.next
        temporaryNode = self.head
        self.head = self.head.next
        temporaryNode.next = self.head
        lastNode.next = self.head
        del temporaryNode
    def deleteAt(self, position):
        currentNode = self.head
        currentPos = 0
        while True:
            if currentPos == position:
                previousNode.next = currentNode.next
                currentNode.next = self.head
                break
            previousNode = currentNode
            currentNode = currentNode.next
            currentPos += 1
    def deleteEnd(self):
        lastNode = self.head
        while lastNode.next is not self.head:
            previousNode = lastNode
            lastNode = lastNode.next
        previousNode.next = lastNode.next
    def printList(self):
        if self.head is None:
            print('list is empty')
            return
        currentNode = self.head
        while True:
            print(currentNode.data)
            currentNode = currentNode.next
            if currentNode is self.head:
                break
        print(currentNode.data)
linkedList = CircularSinglyLinkedList()
node1 = Node('sai')
node2 = Node('krish')
node3 = Node('sushma')
linkedList.insertEnd(node1)
linkedList.insertEnd(node2)
linkedList.insertEnd(node3)
linkedList.deleteAt(1)
linkedList.printList()
```

## Insertion Sort

```python
def insertion_sort(lst):
    for i in range(1, len(lst)):
        cur_ele = lst[i]
        pos = i
        while pos >=1 and cur_ele < lst[pos-1]:
            lst[pos] = lst[pos-1]
            pos -= 1
        lst[pos] = cur_ele
    return lst
lst = [9,5,8,7,1,2]
insertion_sort(lst)
```

## Shell_Sort

```python
def shell_sort(lst):
    gap = len(lst) // 2
    while gap > 0:
        for i in range(gap, len(lst)):
            cur_ele = lst[i]
            pos = i
            while pos >=gap and cur_ele < lst[pos - gap]:
                lst[pos] = lst[pos - gap]
                pos -= gap
            lst[pos] = cur_ele
        gap = gap // 2
    return lst
lst = [9,3,4,7,1,2]
shell_sort(lst)
```
